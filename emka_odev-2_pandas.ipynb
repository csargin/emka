```python
import numpy as np
import pandas as pd
```

```python
"""
  1) İndexlerinde "Breaking Bad" , "Chernobyl" , "Leyla ile Mecnun" , "X-Men" , "Captain America" , "Disaster Movie" olan Datasında 9.4 , 9.3 , 8.8 , 8.4 , 3.8 , 2.1 olan bir seri oluşturun.
  (2 Farklı yolla da oluşturun) (Datayı ve İndexleri liste olarak göndererek ve doğrudan dictionary göndererek.)
"""
dizi_index = np.array(["Breaking Bad" , "Chernobyl" , "Leyla ile Mecnun" , "X-Men" , "Captain America" , "Disaster Movie"])
dizi_data = np.array([9.4 , 9.3 , 8.8 , 8.4 , 3.8 , 2.1])
dizi_columns = np.array(["Rating"])
dizi = pd.DataFrame(data = dizi_data , index = dizi_index , columns = dizi_columns)
print(dizi)
print("\n")

dizi_dict = {}
for key, value in dizi.iterrows():
  dizi_dict[key] = value[0]

print(dizi_dict)
print("\n")

dizi = pd.Series(data = dizi_dict)
print(dizi)
```

```python
"""
  2) donem1_mat adında bir seri oluşturun. Bu serinin içinde "Ali" ,"Deniz" ,"Aslı" ,"Burak" ,"Kerem" ,"Yasemin" ,"Derya" index değerleri olsun, 85, 80 , 94 , 74 , 91 , 96 , 70 data değerleri olsun.
  Bu şekilde donem1 adında öğrencilerin 1.dönem matematik nor ortalamalarının tutulduğu bir seri oluşturun. 
  donem2_mat adında başka bir seri oluşturun. Okula yeni bir öğrenci gelmiş olsun ve indexlerde bu sefer "Ali" ,"Deniz" ,"Aslı" ,"Burak" ,"Kerem" ,"Yasemin" ,"Derya" ,"Mert" indexleri olsun.
  80, 92 , 85 , 87, 90 , 92, 100, 83 ise dataları olsun. Bu iki seriyi birleştirerek bütün öğrencilerin yıl sonu matematik ORTALAMASINI hesaplayın.
  Not : Eğer bir öğrencinin sadece ikinci dönem notu bulunuyorsa, birinci dönem hesaba eklenmeden ikinci dönem notu yıl sonu notu olarak hesaplansın.
"""

donem1_mat_index = np.array(["Ali" ,"Deniz" ,"Aslı" ,"Burak" ,"Kerem" ,"Yasemin" ,"Derya"])
donem1_mat_data = np.array([85, 80 , 94 , 74 , 91 , 96 , 70])

donem2_mat_index = np.array(["Ali" ,"Deniz" ,"Aslı" ,"Burak" ,"Kerem" ,"Yasemin" ,"Derya" ,"Mert"])
donem2_mat_data = np.array([80, 92 , 85 , 87, 90 , 92, 100, 83])

donem1_mat = pd.Series(data = donem1_mat_data , index = donem1_mat_index)
donem2_mat = pd.Series(data = donem2_mat_data , index = donem2_mat_index)

donem_mat = pd.concat([donem1_mat,donem2_mat] , join="outer", sort=True , axis=1 , verify_integrity = True , )
donem_mat.columns = ["Dönem 1", "Dönem 2"]
donem_mat = donem_mat.reindex(columns=["Dönem 1", "Dönem 2", "Ortalama"])

#donem_mat.insert(loc = 2, column = "Ortalama" , value = donem_mat.mean(axis=1, numeric_only=True)) / Kısa Çözüm

if donem_mat["Dönem 1"].isnull().values.any():
  df_null = donem_mat.loc[donem_mat["Dönem 1"].isnull()].copy()
  for i in df_null.index:
    donem_mat.at[i, "Dönem 1"] = donem_mat.at[i, "Dönem 2"]

donem_mat["Ortalama"] = (donem_mat["Dönem 1"] + donem_mat["Dönem 2"]) / 2

print(donem_mat)
```

```python
"""
  3) İndexlerinde "Osman", "Ozlem", "İnci", "Koray", "Melek" olan, columnlarında "Sınıf" , "Cinsiyet" , "Favori Ders" isimlerinde 3 column bulunan bir DF oluşturacaksınız. Osman'ın sınıfı 8 , cinsiyeti "E" ,
  Favori Dersi "Fen" Ozlem'in sınıfı 11 , cinsiyeti "K" , Favori Dersi "Mat" İnci'nin sınıfı 5 , cinsiyeti "K" , Favori Dersi "Sosyal" Koray'ın sınıfı 9 , cinsiyeti "E" , Favori Dersi "Mat" Meleğin sınıfı 7 ,
  cinsiyeti "K" , Favori Dersi "Ing" Olacak şekilde bir df oluşturun.
"""

df_columns = np.array(["Sınıf" , "Cinsiyet" , "Favori Ders"])
df_data = {"Osman": [8,"E", "Fen"],
           "Ozlem": [11,"K", "Mat"],
           "İnci": [5,"K", "Sosyal"],
           "Koray": [9,"E", "Mat"],
           "Melek": [7,"K", "Ing"], 
          }

df = pd.DataFrame.from_dict(data = df_data , orient='index' , columns = df_columns)
print(df)
```

```python
import numpy as np
import pandas as pd

url = 'https://github.com/csargin/Ders-Notebooklari/blob/main/Pandas-Dersleri/tips.csv?raw=true'
df_tips =  pd.read_csv(filepath_or_buffer = url , on_bad_lines='skip')
```

```python
#1) df_tips içindeki Payment ID column'unun son 3 rakamını çekin. Bu işlemi 3 farklı yolla da yapmaya çalışın.

print(df_tips["Payment ID"].tail(3))
print("\n")
print(df_tips["Payment ID"][-3:])
print("\n")
print(df_tips["Payment ID"].iloc[-3:])
print("\n")
```

```python
"""
  2) df_tips içindeki Payer Name sütununda isim ve soy isim bilgileri bulunuyor. O sütundan sadece isim bilgileri çekin ve 'JustName' adında yeni bir column oluşturun.
  (Bu biraz zorlayıcı olabilir, Önce kendiniz yapmaya çalışın fakat yapamazsanız ChatGPT'ye soruyu düzgünce sorarak cevabını kod olarak alabilirsiniz, ardından döndürülen kodu anlamaya çalışmanız yeterlidir.)
"""

df_tips['JustName'] = df_tips['Payer Name'].str.split().str[0]
print(df_tips)
```

```python
#3) 2.Soruda yeni oluşturduğunuz 'JustName' sütunundaki verilerin her birinden kaç tane olduğuna uygun methodu kullanarak bakın. 
df_tips['JustName'].value_counts()
```

```python
"""
  4) 3.Soruda gördüğünüz gibi JustName sütununda 'Jason' isminden 9 tane, 'Michael' isminden 7 tane , 'James' isminden 6 tane bulunuyor.
  Peki bunların hepsi gerçekten farklı kişiler mi ? Hemen 'Payer Name' sütunundaki verilerin adetlerini de saydırtarak bir kıyaslama yapın.
  'JustName' sütununu kullanmanın faydasız ve yanıltıcı olabileceğini görün ardından oluşturduğumuz 'JustName' sütununu kalıcı olarak drop edin. (Oluşturmanız için bir pratikti bu sütun sadece :D)
"""

df_tips['Payer Name'].value_counts()
df_tips.drop('JustName', axis = 1, inplace = True)
print(df_tips)
```

```python
"""
  5) Erkeklerin ve kadınların davranış patternlerini ayrı ayrı inceleyip yorumda bulunabilmek için 'df_male' ve 'df_female' adında 2 dataframe oluşturun.
  Birinde sadece 157 erkek verisi, diğerinde sadece 87 kadın verisi bulunsun.
"""

# Filter for male customers
df_male = df_tips[df_tips['sex'] == 'Male'].sample(n=157, random_state=42)
df_male = df_tips[df_tips['sex'] == 'Male']

# Filter for female customers
df_female = df_tips[df_tips['sex'] == 'Female'].sample(n=87, random_state=42)
df_female = df_tips[df_tips['sex'] == 'Female']

print(df_male.shape)
print(df_female.shape)
```

```python
#6) Ardından bu 'df_male' ile 'df_female' 'in ödedikleri bahşişler toplamını bulup karşılaştırın.

print(df_male["tip"].sum())
print(df_female["tip"].sum())
```

```python
"""
  7) 6.Soruda erkeklerin ve kadınların ödedikleri bahşişler toplamına bakıldığında Erkeklerin kadınlara kıyasla toplamda daha çok bahşiş ödediğini görüyoruz.
  Fakat biz biliyoruz ki bizim verimizde 157 Erkek, 87 kadın verisi vardı. Dolayısıyla Erkek sayısı daha fazla olduğu için toplam ödeyecekleri bahşiş miktarlarında da daha fazla çıkması oldukça normal bir durum.
  Bu yüzden daha mantıklı bir çıkarım için bu sefer df_male ile df_female'in bahşiş ORTALAMALARIN'a bakıp kıyaslayın, aralarında anlamlı ve ciddi bir fark var mı ?
"""

print(df_male["tip"].mean())
print(df_female["tip"].mean())
```

```python
"""
  8) Oluşturmuş olduğunuz df_male ile çalışarak elimizdeki verideki erkeklerin yüzde kaçının sigara içtiğini bulun.
  Aynı şekilde elimizdeki kadınların yüzde kaçının sigara içtiğini de bulun. Aralarında anlamlı bir fark mı yorumlayın.
"""

# Calculate the percentage of male customers who smoke
male_smokers = df_male['smoker'].value_counts(normalize=True)['Yes'] * 100

# Calculate the percentage of female customers who smoke
female_smokers = df_female['smoker'].value_counts(normalize=True)['Yes'] * 100

print("Percentage of male customers who smoke:", male_smokers)
print("Percentage of female customers who smoke:", female_smokers)

# Create contingency table for male and female smokers
contingency_table = pd.crosstab(df_tips['sex'], df_tips['smoker'])
print("contingency_table:", contingency_table)

# Perform chi-squared test on the contingency table
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(contingency_table)

# print the results
print("Chi-squared value:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)

"""
  Assuming a significance level of 0.05, if the p-value is less than 0.05, we can reject the null hypothesis that there is no difference in smoking rates between male and female customers,
  and conclude that there is a statistically significant difference. On the other hand, if the p-value is greater than 0.05, we cannot reject the null hypothesis and conclude that there is not a statistically significant difference.
"""
```
